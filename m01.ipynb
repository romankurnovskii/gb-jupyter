парабола

```python
plt.plot(x,y)
plt.title("Парабола")
```

окружность

```python
a = []
b = []
r = 50
x = np.linspace(-r, r, 250)
for i in x:
    a.append(math.sqrt(r ** 2 - i ** 2))
    b.append(-math.sqrt(r ** 2 - i ** 2))
plt.figure(figsize=(5, 5))
plt.plot(x,a)
plt.plot(x,b)
```

эллипс

```python
a = 10
b = 50
y1 = []
y2 = []
x = np.linspace(-a, a, 100)
for i in x:
    y1.append(b / a * math.sqrt(a ** 2 - i ** 2))
    y2.append(-b / a * math.sqrt(a ** 2 - i ** 2))
plt.plot(x, y1)
plt.plot(x, y2)
```

гипербола

```python
x = np.linspace(-5, 5, 100)
y = 1 / x
plt.plot(x, y)
```

Нарисуйте трехмерный график двух параллельных плоскостей.

Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-1,1,10)
y = np.linspace(-1,1,10)
X,Y = np.meshgrid(x,y)
Z= 4*X + 8*Y + 1

X1,Y1 = np.meshgrid(x,y)
Z1= 4*X1 + 8*Y1 + 10

fig = plt.figure()

ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z)
ax.plot_surface(X1, Y1, Z1)

plt.show()
```

```python
x = np.linspace(-50, 50, 200)
a = 5
b = 5
y = np.sqrt((1+x**2/a**2)*b**2)

X,Y = np.meshgrid(x,y)

Z = X**2/a**2 + Y**2/b**2

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z)
plt.show()
```
